```python
import whois

def get_domain_phone(domain_name):
    try:
        w = whois.whois(domain_name)
        if w.status:
            contacts = w.contacts
        return contacts
    except whois.parser.PywhoisError:
        return None


```

```python
import whois

def get_registrant_contact(domain_name):
    try:
        w = whois.whois(domain_name)
        if w.status:
            registrant_contact = w.registrant
            return registrant_contact
    except whois.parser.PywhoisError:
        return None

# Example usage:
domain = "jsacarbon.com"
registrant = get_registrant_contact(domain)
if registrant:
    print("Registrant Contact:")
    print("Name:", registrant.name)
    print("Email:", registrant.email)
    print("Phone:", registrant.phone)
    # Add more attributes as needed
else:
    print("Registrant contact not found.")
```

```python
import whois

def get_raw_whois(domain):
    w = whois.whois(domain)
    return w.text

# Example usage
domain = "jsacarbon.com"
raw_whois = get_raw_whois(domain)
print(raw_whois)
```

```python
import whois
import re

def get_registrant_phone(domain):
    w = whois.whois(domain)
    raw_whois = w.text
    phone_regex = r"Registrant Phone:\s+(.+)"
    match = re.search(phone_regex, raw_whois, re.IGNORECASE)
    if match:
        return match.group(1)
    else:
        return None

# Example usage
domain = "jsacarbon.com"
registrant_phone = get_registrant_phone(domain)
if registrant_phone:
    print("Registrant Phone:", registrant_phone)
else:
    print("Registrant Phone not found.")
```
