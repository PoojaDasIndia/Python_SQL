```python
import whois
import time

def check_domain_availability(domains, batch_size=100, delay=2):
    available_domains = []

    for i in range(0, len(domains), batch_size):
        batch = domains[i:i+batch_size]

        for domain in batch:
            try:
                w = whois.whois(domain)
                if w.status:
                    print(f"Domain '{domain}' is taken")
                else:
                    print(f"Domain '{domain}' is available")
                    available_domains.append(domain)
            except Exception as e:
                print(f"Domain '{domain}' is available")
                # print(f"Error checking domain '{domain}': {str(e)}")

        time.sleep(delay)

    return available_domains

# Example usage
# domain_list = ["example1.com", "example2.com", "example3.com", ...]  # List of 5000 domains
# available_domains = check_domain_availability(domain_list)
# print("Available domains:")
# print(available_domains)
bcguctadbqytskjt
```
